RandInt
-----------
- randint internally called randrange,
- 0-10 include 10 also
- Return random integer in range [a, b], including both end points.


```python
from random import randint,randrange

for _ in range(0,100):
    ran = randint(0,10)
    print(ran,end=',')
```

RandRange
-----------------
- Roughly equivalent to choice(range(start, stop, step)) but supports arbitrarily large ranges and is optimized for common cases.

```python
from random import randint,randrange

for _ in range(0,100):
    ran = randrange(0,10)
    print(ran,end=',')
```

Choice
-------
Choose a random element from a non-empty list

```python
from random import choice

random_name = ["ashish","sreshti","bindra","mui"]

for _ in range(1,4):
    print(choice(random_name))
```

 choices
 -----
 
 Syntax : random.choices(sequence, weights=None, cum_weights=None, k=1) 

```python
from random import choices

random_name = ["ashish","sreshti","bindra","mui"]

for _ in range(3):
    print(choices(random_name))
```

```python
from random import choices

random_name = ["ashish","sreshti","bindra","mui"]

for _ in range(3):
    print(choices(random_name,k=2))
```

```python
from random import choices

random_name = ["ashish","sreshti","bindra","mui"]

for _ in range(3):
    print(choices(random_name,weights=[2,3,1,1],k=2))
# weights 2 for sreshti meanse sreshti will come more likly to be first
```

random
--------
Parameters : This method does not accept any parameter.

Returns : This method returns a random floating number between 0 and 1.

```python
from random import random

for _ in range(10):
    print(random())


print(f'{"end":*^20}')

for _ in range(10):
    ten = random()*100
    print(f'{ten:.2f}')
```
