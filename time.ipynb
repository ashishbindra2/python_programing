```python
from timeit import timeit

def normal_impl() -> list[int]:
    return[i for i in (0,1,2,3,4)]

def optimised_impl() -> list[int]:
    l =[]
    for i in (0,1,2,3,4):
        l.append(i)
    return l

normal_time: float = timeit(stmt=normal_impl)
optimised_time: float = timeit(stmt=optimised_impl)
    
print(f'normal_impl: {round(normal_time, 4)}s')
print(f'optimised_impl: {round(optimised_time, 4)}s')
```

```python
# example
import timeit
t= timeit.timeit('print("hello")',number=1000)
print(f'The time taken {t} sec')
```

```python
from datetime import datetime

# Define your date string
date_string = "11-02-2024 02:26:25"

# Convert the date string to a datetime object
date_object = datetime.strptime(date_string, "%d-%m-%Y %H:%M:%S")

# Convert the datetime object to a timestamp
timestamp = date_object.timestamp()

print("Timestamp:", timestamp)
```

```python
import time
def timeing_with_time():
    start = time.time()
    time.sleep(1)
    end = time.time()
    print(end - start)
    
timeing_with_time()
```

```python
import time
def timeing_with_time():
    start = time.perf_counter()
    time.sleep(1)
    end = time.perf_counter()
    print(end - start)
    
timeing_with_time()
```
