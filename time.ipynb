```python
from timeit import timeit

def normal_impl() -> list[int]:
    return[i for i in (0,1,2,3,4)]

def optimised_impl() -> list[int]:
    l =[]
    for i in (0,1,2,3,4):
        l.append(i)
    return l

normal_time: float = timeit(stmt=normal_impl)
optimised_time: float = timeit(stmt=optimised_impl)
    
print(f'normal_impl: {round(normal_time, 4)}s')
print(f'optimised_impl: {round(optimised_time, 4)}s')
```

```python
class Calculator:
    def __init__(self,values: list[float]):
        self.values = values
    
    def sum(self) -> float:
        print(f'Getting the sum of: {self.values}')
        return sum(self.values)
    
    def mean(self) -> float:
        print(f'Getting the mean of : {self.values}')
        return mean(self.values)

numbers: list[float] = [5.5, 10.5, 30.5, 32.6, 20.5,9.5]

calc: Calculator = Calculator(numbers)

print(calc.sum())
print(calc.sum())
print(calc.sum())

print(calc.mean())
print(calc.mean())
print(calc.mean())
```

```python
class Calculator:
    def __init__(self,values: list[float]):
        self.values = values
    @cached_property
    def sum(self) -> float:
        print(f'Getting the sum of: {self.values}')
        return sum(self.values)
    @cached_property
    def mean(self) -> float:
        print(f'Getting the mean of : {self.values}')
        return mean(self.values)

numbers: list[float] = [5.5, 10.5, 30.5, 32.6, 20.5,9.5]

calc: Calculator = Calculator(numbers)

print(calc.sum)
print(calc.sum)
print(calc.sum)

print(calc.mean)
print(calc.mean)
print(calc.mean)
```

```python
from timeit import timeit,repeat

list_comp = """list_comp: list[int] = [i for i in range(10)]"""

for_loop = """for_loop : list[int] = []

for i in range(10):
    for_loop.append(i)
"""
list_comp_time: float = min(repeat(list_comp, repeat = 5, number= 1_000_000))
for_loop_time: float = min(repeat(for_loop, repeat=5, number=1_000_000))

print(f'List comp: {list_comp_time:.3f}s')
print(f'for comp: {for_loop_time:.3f}s')
```

```python
list_comp_time: float = timeit(list_comp,  number= 1_000_000)
for_loop_time: float = timeit(for_loop, number=1_000_000)

print(f'List comp: {list_comp_time:.3f}s')
print(f'for comp: {for_loop_time:.3f}s')
```

```python
for_loop_time: float = timeit(for_loop, number=1_000_000)
list_comp_time: float = timeit(list_comp,  number= 1_000_000)


print(f'List comp: {list_comp_time:.3f}s')
print(f'for comp: {for_loop_time:.3f}s')
```

```python
from funtools import cache, lru_cache

@cache # stores all previous values
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

for i in range(400):
# for i in enumerate(400):

    print(i,fib(i))
print("done")
```

```python

@lru_cache(maxsize=5) # least resently used
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

for i in range(400):
    print(i,fib(i))
print("done")
```
