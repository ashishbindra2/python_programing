Pillow is helpful when we have to work with multiple images at a time

To convert all images into .png

To loop over the files in current directory we will use `os` module

```python
#  to list out all files and folders
from os import listdir

for list_file in listdir(paths):
    print(list_file)
```

```python
# program to list out .jpg file in folder
from os import listdir

for list_file in listdir(paths):
    if list_file.endswith('.jpg'):
        print(list_file)
        
```

```python
#  program to print file name and extention
from os import listdir
from os.path import splitext
for list_file in listdir(paths):
    if list_file.endswith('.jpg'):
        data = splitext(list_file)
        print(data)
```

```python
# To convert all images into .png
from PIL import Image
from os.path import splitext

for list_file in listdir(paths):
    if list_file.endswith('.jpg'):
        
        image = Image.open(paths+list_file)
        file_name, extention = splitext(list_file)
        image.save(f"{paths}pngs/{file_name}.png")
```

```python
# resize image for thumbnail or gallary
from PIL import Image
from os.path import splitext

size_300 = (300, 300)
size_700 = (700, 700)


for list_file in listdir(paths):
    if list_file.endswith('.jpg'):
        image = Image.open(paths+list_file)
        file_name, extention = splitext(list_file)
                
        image.thumbnail(size_700)
        image.save(f"{paths}700/{file_name}_700{extention}")
        
        image.thumbnail(size_300)
        image.save(f"{paths}300/{file_name}_300{extention}")
```

Blendd images

```python
from PIL import Image

# Open the images
image_1 = Image.open('test/img/Pillow/1_blur_10.jpg')
image_2 = Image.open('test/img/Pillow/1_b&w.jpg')


# Ensure both images are in the same mode
image_1 = image_1.convert("RGB")
image_2 = image_2.convert("RGB")

# Resize the second image to match the first image's size
image_2 = image_2.resize(image_1.size)

# Blend the images with equal weight
blended_image = Image.blend(image_1, image_2, alpha=0.5)

# Show the result
blended_image.show()
```

composite images

```python
from PIL import Image, ImageFilter

# Open the images
image_1 = Image.open('test/img/Pillow/1_blur_10.jpg')
image_2 = Image.open('test/img/Pillow/1_b&w.jpg')

# Ensure both images are in the same mode and have the same size
image_1 = image_1.convert("RGB")
image_2 = image_2.convert("RGB")
image_2 = image_2.resize(image_1.size)

# Create a mask image
mask = Image.open('test/img/Pillow/1_blur_10.jpg').convert("L")  # Convert to grayscale for the mask

# Composite the images using the mask
composite_image = Image.composite(image_1, image_2, mask)

# Show the result
composite_image.show()
```

```python
"""Resizes an image and keeps aspect ratio. Set mywidth to 
the desired with in pixels."""

import PIL
from PIL import Image

mywidth = 400

img = Image.open('test/img/Pillow/1.jpg')
wpercent = (mywidth/float(img.size[0]))
hsize = int((float(img.size[1])*float(wpercent)))
img = img.resize((mywidth,hsize))
img.save('test/img/Pillow/resized.jpg')
```

```python
from PIL import Image, ImageOps
size = (300, 450)
with Image.open("test/img/Pillow/1.jpg") as im:
    ImageOps.contain(im, size).save("test/img/Pillow/imageops_contain.png")
#     ImageOps.cover(im, size).save("test/img/Pillow/imageops_cover.png")
    ImageOps.fit(im, size).save("test/img/Pillow/imageops_fit.png")
    ImageOps.pad(im, size, color="#f00").save("test/img/Pillow/imageops_pad.png")

    # thumbnail() can also be used,
    # but will modify the image object in place
    im.thumbnail(size)
    im.save("test/img/Pillow/imageops_thumbnail.png")
```
