## Thread
- A flow of execution
- An independent part of some program
- for every thread separate job is defined
- The main purpose of that thread is to execute corresponding job.
- Once job execution completes, automatically the tread will be terminated.

### define/create/start a trad for some job
1. Creating a thread without using any class
    - 1st job execution by main thread || sequence exceution
    - 2nd job execution by main thread || sequence exceution
    these are called sequence exceution
    

```python
from threading import Thread
def display():
    for i in range(10):
        print('A')

t = Thread(target = display) # creating a thread to execute display() function
t.start() # Start of a tread

for i in range(10):
    print('B')
```

```python
import time

def doubles(numbers):
    for n in numbers:
        time.sleep(1)
        print("double Value:", 2*n)

def squares(numbers):
    for n in numbers:
        time.sleep(1)
        print("Square Value:", n*n)

numbers = [1,2,3,4,5,6,7,8]

begintime = time.time()

doubles(numbers)
squares(numbers)

endtime =  time.time()

print("The total Time taken:", endtime-begintime)
```

```python
import time
from threading import *

def doubles(numbers):
    for n in numbers:
        time.sleep(1)
        print("double Value:", 2*n)

def squares(numbers):
    for n in numbers:
        time.sleep(1)
        print("Square Value:", n*n)

numbers = [1,2,3,4,5,6,7,8]

begintime = time.time()

t1 = Thread(target = doubles, args = (numbers,))
t2 = Thread(target = squares, args = (numbers,))

t1.start()
t2.start()

t1.join() # means main() tread waiting until completing t1 thread
t2.join()

endtime =  time.time()

print("The total Time taken:", endtime-begintime)
```
